# Desafio

Indicar os **10 maiores clientes por total de venda**.

Para a equipe de vendas elaborar estratégias para aumentar as vendas.

##Para encontrar os principais clientes por volume de vendas:


#####['vendas'].sum() = Soma a coluna vendas para cada grupo.
#####.nlargest(10) = Obtem os 10 principais resultados por volume de vendas.


```python
df_cliente_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
df_cliente_vendas = df_cliente_vendas.reset_index()
df_cliente_vendas
```

####Estilizando a tabela:

```python
df_cliente_vendas.columns = ['Clientes','Vendas']
df_cliente_vendas['Ranque'] = df_cliente_vendas.index + 1
df_cliente_vendas.set_index('Ranque', inplace = True)
df_cliente_vendas
```

###Criando objeto style para representação estilizada do DataFrame

```python
s = df_cliente_vendas.style
s
```

####Alterando a formatação dos dados vendas

```python
s.format({'Vendas':'R$ {:,.2f}'})
s
```

# Desafio 3

```python
df_regiao.index.name = 'Regiao'
```

```python
import numpy as np
```

```python
df_regiao['Porcentagem regiao'] = porcentagem
df_regiao
```

###Configurando o estilo

```python
estilo_regiao = df_regiao.style
estilo_regiao
```

# Desafio 4

Identificar o padrão de **quantidade de produtos solicitados por mês por departamento**.

```python
df_meses = df.copy()
df_meses = df_meses.sort_values('data_pedido')
df_meses
```

###Alterando formatação da data:

%Y em letra maiúscula indicando o ano ("year")

espaço + traço + espaço

%b em letra minúscula para colocar as três primeiras letras do nome do mês em extenso

```python
df_meses['Meses'] = df_meses['data_pedido'].dt.strftime('%Y - %b')
```

###Resetando index da tabela:

```python
df_meses = df_meses.reset_index(drop=True)
```

```python
df_meses
```

####Criação de tabela dinâmica pelo Pandas

```python
vendas_mensais = df_meses.pivot_table(index = 'departamento', columns = 'Meses', values = 'quantidade',
                                      aggfunc = 'sum', sort = False)
vendas_mensais
```

```python
estilo_mensal = vendas_mensais.style
estilo_mensal
```

###Adicionando visualização fixada nas colunas do index

```python
estilo_mensal.set_sticky(axis = 'index')
```

```python
colunas = {
    'selector': '.col_heading',
    'props': 'font-weight: normal; font-family: Arial'
}
tabelas = {
    'selector': 'td,th',
    'props': 'text-align: left;'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: bold; font-family: Arial; text-align: right;'
}
estilo_mensal.set_table_styles([colunas,tabelas,index],overwrite=False)
```

# Desafio 5: Relatório de performance

Construindo uma visualização que permita entender a **relação entre tipos de clientes e modo de envio de produtos de acordo com as vendas**.

```python
df_cliente = df.pivot_table(index = 'segmento_cliente', columns = 'modo_envio', values = 'vendas', aggfunc = 'sum')
df_cliente
```

```python
df_cliente['Total'] = df_cliente.sum(axis=1)
df_cliente.loc['Total'] = df_cliente.sum()
df_cliente
```

####Adicionando o elemento style

```python
compra_cliente = df_cliente.style.format('{:,.2f}')
compra_cliente
```

####Adicionando estilos:

```python
tabela = {
    #exclui elementos com a classe "index_name" da seleção.
    #td são células do cabeçalho. th são células da tabela.
    'selector': 'td,th:not(.index_name)',
    'props': 'font-weight: normal; font-family: Arial; text-align: center; background-color: white'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: normal; text-align: rigth; font-style: italic; color: #696969'
}
compra_cliente.set_table_styles([tabela,index])
```

###Destacando Linhas com style


```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    {
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }],
    'B2B': [{
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    {
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }]
}, overwrite = False, axis=1)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #D8D8D8'
    }]
}, overwrite=False, axis=0)
```

```python
cores_coluna = pd.DataFrame(['false', 'true', 'false'], index= df_cliente['Total'].index, columns=['Total'])
compra_cliente.set_td_classes(cores_coluna)
```
